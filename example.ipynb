```python
import exam

# === Ввод текста пользователем ===
print("Введите текст на английском (до 200 слов):")
lines = []
while True:
    line = input().strip()
    if not line and lines:
        break
    elif not line:
        print("⚠️ Текст не может быть пустым.")
    else:
        lines.append(line)

original_text = " ".join(lines)
if len(original_text.split()) > 200:
    original_text = " ".join(original_text.split()[:200])
    print("⚠️ Обрезаем до 200 слов")

# === Перевод и оценка ===
evaluator = exam.TranslationEvaluator()
translations, scores = evaluator.evaluate_translations(original_text)

# === Вывод результатов ===
for model_name in translations:
    print(f"\n--- {model_name} ---\n{translations[model_name]}")
    print(f"BLEU: {scores[model_name]:.2f}")
```
